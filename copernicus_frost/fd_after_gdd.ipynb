```python
# =======
# function cumulative GDD until last frost occurence

def bedd_until_last_fd(bedd_data, fd_data):
    """
    Accumulated growing degree days (GDD) from January 1 to the date of the last frost occurrence
    """
    # define output lists
    bedd_sum = []
    fd_last = []
    
    # loop trough years
    for year in np.unique(fd_data['time.year']):
        
        # select current year and only first 7 months
        fd_year = fd_data.sel(time = fd_data.time.dt.year.isin(year) & fd_data.time.dt.month.isin([1,2,3,4,5,6,7]))
        
        
        # find index of last FD
        ind = np.max(np.nonzero(fd_year.values))
        fd_last.append(ind)
        #fd_last.append(fd_year['time'][(ind)].values)
        
        # cumulative GDD until last FD
        bedd_sum.append(bedd_data.sel(time=bedd_data.time.dt.year.isin(year) & bedd_data.time.dt.month.isin([1,2,3,4,5,6,7]))[:(ind+1)].sum(dim="time").values)
        
    return bedd_sum, fd_last  


# =======
# not used
def fd_after_bedd(bedd_data, fd_data, threshold):
    """
    looks for frost day after growing degree days reach threshold
    
    input: time series of fd and bedd at specific location
    output: if fd after bedd:
                date 
            else: zero
    """
    
    date_bedd = []
    date_fd = []
    
    # loop trough years
    for year in np.unique(bedd_data['time.year']):
        
        # cumulative sum of bedd for current year year
        bedd_cumul = bedd_data.sel(time=bedd_data.time.dt.year.isin(year) & bedd_data.time.dt.month.isin([1,2,3,4,5,6,7])).cumsum(dim="time")
        fd_cumul = fd_data.sel(time = fd_data.time.dt.year.isin(year) & fd_data.time.dt.month.isin([1,2,3,4,5,6,7]))
        
        # find index and date of bedd_cumul > threshold
        ind = np.argmax(bedd_cumul > threshold)
        date_bedd.append(bedd_cumul['time'][ind].values)
        
        # is there a frost day following
        check_fd = fd_cumul.values[(ind.values):]
        
        if np.any(check_fd > 0):
            date_fd.append(fd_cumul['time'][(ind.values):].values[np.nonzero(check_fd)])
            
        else:
            date_fd.append(0)
            
    return date_bedd, date_fd
```

```python
# =======
# calculate cumulative GDD and date of last FD

# era
#date_bedd_era, date_fd_era = fd_after_bedd(bedd_era_zurich, fd_era_zurich, 80)
bedd_sum_era, fd_last_era = bedd_until_last_fd(bedd_era_zh, fd_era_zh)

# hist
#date_bedd_hist, date_fd_hist = fd_after_bedd(bedd_hist_zurich, fd_hist_zurich, 80)
bedd_sum_hist_hdgem, fd_last_hist_hdgem = bedd_until_last_fd(bedd_hist_hadgem_zh, fd_hist_hadgem_zh)
bedd_sum_hist_gfdl, fd_last_hist_gfdl = bedd_until_last_fd(bedd_hist_gfdl_zh, fd_hist_gfdl_zh)
bedd_sum_hist_ipsl, fd_last_hist_ipsl = bedd_until_last_fd(bedd_hist_ipsl_zh, fd_hist_ipsl_zh)

# rcp6
#date_bedd_rcp6, date_fd_rcp6 = fd_after_bedd(bedd_rcp6_zurich, fd_rcp6_zurich, 80)
bedd_sum_rcp6_hdgem, fd_last_rcp6_hdgem = bedd_until_last_fd(bedd_rcp6_hadgem_zh, fd_rcp6_hadgem_zh)
bedd_sum_rcp6_gfdl, fd_last_rcp6_gfdl = bedd_until_last_fd(bedd_rcp6_gfdl_zh, fd_rcp6_gfdl_zh)
bedd_sum_rcp6_ipsl, fd_last_rcp6_ipsl = bedd_until_last_fd(bedd_rcp6_ipsl_zh, fd_rcp6_ipsl_zh)
```

```python
# =======
# function for cumulative GDD until last FD for each grid 

def bedd_until_last_fd(bedd_data, fd_data):
    """
    Accumulated growing degree days (GDD) from January 1 to the date of the last frost occurrence
    """
    
    # define output lists
    output_bedd = []
    output_fd = []
    
    # loop trough years
    for year in np.unique(fd_data['time.year']):
        
        print(year)
        
        # select current year
        fd_year = fd_data.sel(time = fd_data.time.dt.year.isin(year) & fd_data.time.dt.month.isin([1,2,3,4,5,6,7]))
        # argmax does not work with nans: define mask where nans and replace nans with integer
        mask = fd_year.isel(time=0).notnull() 
        fd_year = fd_year.fillna(-9999) 
        
        # flag all FD with 1
        ind_last = xr.where(fd_year > 0 , 1, fd_year) 
        # find index of last fd (e.g. index of last 1 for each grid cell)
        ind_last = ind_last[::-1].argmax(dim='time', skipna=True).where(mask)
        # get timesteps between 1st january to last fd
        ind_last = len(fd_year) - ind_last
        # correct for regions where never fd
        ind_last = xr.where(fd_year.sum(dim='time') < 1, 0, ind_last) #.where(mask)
        # append to list
        output_fd.append(ind_last.where(mask))
        
        
        #create 3d array filled with 1 at each location from time = 0 until ind_last
        ind_matrix = np.zeros((21,80,90))
        #loop trough lat
        for yy in np.arange(np.shape(ind_matrix)[1]):
            #loop trough long
            for xx in np.arange(np.shape(ind_matrix)[2]):
                ind_matrix[:ind_last[yy,xx].values.astype(int),yy,xx] = 1
        
        # calculate cumulative bedd between 1st January and last fd for each cell
        bedd_year = bedd_data.sel(time=bedd_data.time.dt.year.isin(year) & bedd_data.time.dt.month.isin([1,2,3,4,5,6,7]))
        bedd_sum = bedd_year.where(ind_matrix==1).where(mask).sum(dim="time")
        # append to output list
        output_bedd.append(bedd_sum.where(mask))
    
    # list to xarray dataarray
    output_bedd = xr.concat(output_bedd, dim='time')
    output_fd = xr.concat(output_fd, dim='time')
    return output_bedd, output_fd  
```

```python
# =======
# calculate 

# era
bedd_sum_era, fd_last_era = bedd_until_last_fd(bedd_era_eu, fd_era_eu)
bedd_sum_era.to_netcdf(path='data/results/bedd_sum_era.nc')
# bedd_sum_era = xr.open_dataset('data/results/bedd_sum_era.nc')

# hist
bedd_sum_hist, fd_last_hist = bedd_until_last_fd(bedd_hist_eu, fd_hist_eu)
bedd_sum_hist.to_netcdf(path='data/results/bedd_sum_hist.nc')
# bedd_sum_hist = xr.open_dataset('data/results/bedd_sum_hist.nc')

# rcp6
bedd_sum_rcp6, fd_last_rcp6 = bedd_until_last_fd(bedd_rcp6_eu, fd_rcp6_eu)
bedd_sum_rcp6.to_netcdf(path='data/results/bedd_sum_rcp6.nc')
# bedd_sum_rcp6 = xr.open_dataset('data/results/bedd_sum_rcp6.nc')
```
